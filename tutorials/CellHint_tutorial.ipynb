## Data

### Lake et al.

```python
adata_Lake0 = sc.read_h5ad("Kidney-2025/Data/Lake-KPMP-2023/adata_normal.h5ad")
```

```python
adata_Lake = adata_Lake0.copy()
adata_Lake = adata_Lake.raw.to_adata()
```

```python
del adata_Lake.var
del adata_Lake.uns
del adata_Lake.obsm
adata_Lake
```

```python
adata_Lake.obs['dataset'] = "Lake et al. (2023) Nature"
adata_Lake.obs['original_annotation'] = adata_Lake.obs['subclass.full']
```

```python
adata_Lake
```

### mBDRC

```python
adata_mBDRC0 = sc.read_h5ad("Kidney-2025/Data/AceraMateos-mBDRC-2025/b7e3c5b0-4f3b-4661-b043-752205b89459.h5ad")
```

```python
adata_mBDRC = adata_mBDRC0.copy()
adata_mBDRC = adata_mBDRC.raw.to_adata()
```

```python
del adata_mBDRC.var
del adata_mBDRC.uns
del adata_mBDRC.obsm
adata_mBDRC
```

```python
adata_mBDRC.obs['dataset'] = "Acera-Mateos et al. (2025) bioRxiv"
adata_mBDRC.obs['original_annotation'] = adata_mBDRC.obs['author_cell_type']
```

```python
adata_mBDRC
```

### concat into one adata object

```python
adata = sc.concat([adata_Lake, adata_mBDRC])
```

```python
adata
```

## CellHint harmonization

#### scanpy steps

```python
sc.pp.normalize_total(adata, target_sum = 1e4)
sc.pp.log1p(adata)
sc.pp.highly_variable_genes(adata, batch_key = 'dataset', subset = True)
sc.pp.scale(adata, max_value = 10)
sc.tl.pca(adata)
sc.pp.neighbors(adata)
```

```python
sc.tl.umap(adata)
```

```python
sc.pl.umap(adata, color = 'dataset')
```

```python
sc.pl.umap(adata, color = 'original_annotation')
```

#### harmonization 

```python
alignment = cellhint.harmonize(adata, 'dataset', 'original_annotation')
```

```python
alignment.reannotation
```

```python
cellhint.treeplot(alignment, figsize=(10,20), save=True)
```

```python
a = alignment.reannotation['reannotation'].unique()
```

```python
import re
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(columns=['mBDRC','Lake'])
for ss in a:
    my_string = ss
    split_list = re.split(r" [=∋∈] ", my_string)
    df.loc[len(df)] = split_list
```

```python
df
```

```python
df_replaced = df.replace(['NONE','UNRESOLVED'], np.nan)
```

```python
df_replaced
```

```python
df_cleaned = df_replaced.dropna()
```

```python
df_cleaned
```

#### save cellhint matches

```python
df_cleaned.to_csv("Kidney-2025/CellHint/cellhint_matches.csv", index=False)
```
